```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('1553768847_housing.xlsx')
df
```

```python
df.head()
```

```python
df_nan = df[df.isna()]
df_nan
```

```python
mean_value=df['latitude'].mean()
mean_value1=df['housing_median_age'].mean()
mean_value2=df['total_rooms'].mean()
mean_value3=df['total_bedrooms'].mean()
mean_value4=df['population'].mean()
mean_value5=df['households'].mean()
mean_value6=df['median_income'].mean()
mode_value=df['ocean_proximity'].mode()
print(mean_value)
print(mean_value1)
print(mean_value2)
print(mean_value3)
print(mean_value4)
print(mean_value5)
print(mean_value6)
print(mode_value)
```

```python
df['latitude'].fillna(value=mean_value, inplace=True)
df['housing_median_age'].fillna(value=mean_value1, inplace=True)
df['total_rooms'].fillna(value=mean_value2, inplace=True)
df['total_bedrooms'].fillna(value=mean_value3, inplace=True)
df['population'].fillna(value=mean_value4, inplace=True)
df['households'].fillna(value=mean_value5, inplace=True)
df['median_income'].fillna(value=mean_value6, inplace=True)
df['ocean_proximity'].fillna(value=mode_value, inplace=True)
df
```

```python
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
from sklearn import preprocessing
feature_cols = ['longitude','latitude','housing_median_age','total_rooms','total_bedrooms','population','households','median_income','ocean_proximity']

X = df[feature_cols]
standard = preprocessing.scale(X)
X
```

```python
y=df.median_house_value
s1=preprocessing.scale(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state= 482739)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
from sklearn import metrics
```

```python
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```
